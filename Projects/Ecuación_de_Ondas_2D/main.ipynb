### Proyecto 07: Ecuación de Ondas 1D

El objetivo de este trabajo es modelar la propagación de ondas en una cuerda unidimensional utilizando el método de diferencias finitas para resolver la ecuación de ondas dependiente del tiempo. Se analizarán diferentes condiciones de frontera e iniciales, así como los efectos de la densidad lineal de la cuerda y la tensión aplicada. También se evaluará la estabilidad numérica del esquema mediante el criterio de Von Neumann, investigando el impacto del tamaño de los pasos espaciales y temporales. Los resultados se presentarán mediante gráficas (y en lo posible, animaciones) que muestren la propagación de las ondas y la comparación de los resultados para diferentes condiciones físicas.

**1. Discretización de la ecuación de ondas:**

La ecuación diferencial es la siguiente:

$$
\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2},
$$

donde $u(x, t)$ representa la desviación de la cuerda, y $c = \sqrt{\frac{T}{\mu}}$ es la velocidad de propagación de la onda, siendo $T$ la tensión en la cuerda y $\mu$ la densidad lineal de masa.  

Utilice *diferencias centrales* (*central difference*) para aproximar las derivadas espaciales y temporales. Proponga el esquema numérico correspondiente.


Aproximamos las derivadas parciales con *central difference*:

$$ \frac{\partial^2 u}{\partial t^2} \approx \frac{1}{(\Delta t)^2} \left(  u_{i, j+1} + u_{i, j-1} - 2u_{i,j} \right)$$

$$ \frac{\partial^2 u}{\partial x^2} \approx \frac{1}{(\Delta x)^2} \left(  u_{i+1, j} + u_{i-1, j} - 2u_{i,j} \right)$$

Ahora reemplazamos en la EDP:

$$ \frac{1}{(\Delta t)^2} \left(  u_{i, j+1} + u_{i, j-1} - 2u_{i,j} \right) = \frac{c^2}{(\Delta x)^2} \left(  u_{i+1, j} + u_{i-1, j} - 2u_{i,j} \right)$$

Despejamos el término $j+1$:

$$ \boxed{ \textcolor{blue}{
    u_{i, j+1} =   \left( \frac{c \Delta t}{\Delta x} \right)^2 \left(  u_{i+1, j} + u_{i-1, j} - 2u_{i,j} \right) - u_{i, j-1} + 2u_{i,j}
}}$$
